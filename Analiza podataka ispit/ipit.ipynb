```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
df_nums = df.drop(columns=['fueltype', 'aspiration', 'doornumber', 'carbody', 'drivewheel', 'enginelocation', 'enginetype', 'cylindernumber', 'fuelsystem', 'CarBrand'])
df_nums
```

```python
corr_mat = df_nums.corr()['price']
corr_mat
```

Kolone curbweight, enginesize i horsepower su u najvecoj korelaciji sa kolonom price

Kolone peakrpm, citympg, highwaympg, stroke i compressionratio nisu u skoro nikakvoj korelaciji sa cijenom

```python
X = df['enginesize'].values.reshape(-1,1)
y = df['price'].values
```

```python
sns.regplot(data=df, x='enginesize', y='price')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.predict([[150]])[0].round(2)
```

```python
y_train_pred = lr.predict(X_train)
y_test_pred = lr.predict(X_test)
```

```python
print('Mean sum of squares (MSE): %.2f ' % mean_squared_error(y_test, y_test_pred))
print('R2-score: %.2f' % r2_score(y_test, y_test_pred) )
```
