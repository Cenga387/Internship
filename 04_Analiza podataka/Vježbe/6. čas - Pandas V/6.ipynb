### GroupBy

```python
data = {'Ime': ['Ajla', 'Iman', 'Haris', 'Ajdin', 'Mia', 'Sara'],
        'DobnaSkupina': ['3-5', '6-8', '3-5', '6-8', '3-5', '6-8'],
        'Igračke': [10, 5, 8, 12, 15, 7]}
```

```python
df = pd.DataFrame(data)
```

```python
grupirano_po_dobnoj_skupini = df.groupby('DobnaSkupina')['Igračke'].sum()
```

Iteracija kroz grupe

```python
sektori = df.groupby("Sector")
df_1 = pd.DataFrame(columns= df.columns)
df_1
```

```python
for sektor, data in sektori:
    highest_revenue_company = data.nlargest(1, "Revenue")
    df_1 = pd.concat([df_1, highest_revenue_company])
```

```python
df_1
```

```python
gradovi = df.groupby("Location")
df_1 = pd.DataFrame(columns= df.columns)
df_1
```

```python
for grad, data in gradovi:
    highest_revenue_grad = data.nlargest(1, "Revenue")
    df_1 = pd.concat([df_1, highest_revenue_grad])
```

```python
df_1
```
